```python
import mysql.connector
c = mysql.connector.connect(user="root", password = "", database = "dbmslab", host = '127.0.0.1')
cursor = c.cursor()
cursor.execute(''' CREATE Table IF NOT EXISTS Person(
               ID INT AUTO_INCREMENT PRIMARY KEY,
               Name varchar(255),
               Contact_No INT,
               Address varchar(255));''')

def insert_data():
    name = input("enter your name: ")
    Contact_No = int(input("enter your Contact Number: "))
    Address = input("enter your Address: ")
    data = []
    data.append(name)
    data.append(Contact_No)
    data.append(Address)
    query = "INSERT INTO Person (Name, Contact_No, Address) values (%s, %s, %s);"
    cursor.execute(query, (data[0], data[1], data[2]))
    c.commit()
    print("data added successfully!\n")
    print("The added data is: ")
    cursor.execute(f"select * from Person where Name = '{name}'")
    results = cursor.fetchall()
    for r in results:
        print(r)

def update_data():
    ID = input("enter the ID of the person you want to update: ")
    if ID=="":
        print("ID is empty, please enter the correct ID")
    else:
        name = input("enter the new name of the person (leave blank if you do not want to update): ")
        Address = input("enter the new Address of the person (leave blank if you do not want to update): ")
        Contact_No = input("enter the new contact no of the person (leave blank if you do not want to update): ")
    #check if the given ID exists in the database
        cursor.execute(f"select * from Person Where ID = {ID}")
        result = cursor.fetchall()
        if result:
            query = "update Person set "
            if name:
                query+= f"Name = '{name}',"
            if Address:
                query+= f"Address = '{Address}',"
            if Contact_No:
                query+= f"Contact_No = '{Contact_No}',"
            query = query.rstrip(", ")
            query += f" WHERE id = {ID};"
            print(query)
            cursor.execute(query)
            c.commit()
            print("Person updated successfully!\n")
            print("The updated data is: \n")
            cursor.execute(f"select * from Person Where ID = {ID}")
            result = cursor.fetchall()
            for r in result:
                print(r)
        else:
            print("The given ID does not exist in the Person Database.")
    
def view_data():
    cursor.execute("select * from Person")
    results = cursor.fetchall()
    for r in results:
        print(r)
        
while(True):
    print(" 1: Insert Data")
    print(" 2: Update Data")
    print(" 3: View Data")
    print(" 4: Quit")
    
    choice = int(input("Enter the choice: "))
    if choice == 1:
        insert_data()
    elif choice == 2:
        update_data()
    elif choice == 3:
        view_data()
    elif choice == 4:
        break
    else:
        print("invalid choice ")

cursor.close()
c.close()

    
    
```
